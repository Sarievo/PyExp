**Encryption/Decryption:**

**Encrypting and decrypting**

Use `openssl genrsa -out private_key.pem 2048` to generate a 2048-bit RSA private key. use `cat` to print it to the screen.

Use `openssl rsa -in private_key.pem -outform PEM -pubout -out public_key.pem` to create a public key based on a private key.

Use `openssl rsautl -encrypt -pubin -inkey public_key.pem -in secret.txt -out secret.enc` to encrypt the file using your public key. This creates the file "secret.enc".

Use `openssl rsautl -decrypt -inkey private_key.pem -in secret.enc` to print the decrypted content to the screen.

**Creating a hash digest**

Use `openssl dgst -sha256 -sign private_key.pem -out secret.txt.sha256 secret.txt` to create a hash digest of the message. Use `openssl dgst -sha256 -verify public_key.pem -signature secret.txt.sha256 secret.txt` to perform verification.

**Hashing (MD5)**

Use `md5sum file.txt > file.txt.md5` to generate the MD5 sum for the file and store it. Use `md5sum -c file.txt.md5` to verify that the hash is correct, and that the original file hasn't been tampered with since the sum was made. Use `cp a.txt b.txt` to copy a to b.

**Hashing (SHA1)**

Use `shasum file.txt > file.txt.sha1` to generate the SHA1 sum for the file and store it. Use `shasum -c file.txt.sha1` to verify that the hash is correct, and that the original file hasn't been tampered with since the sum was made.

**Hashing (SHA256)**

Use `shasum -a 256 file.txt > file.txt.sha256` to generate the SHA1 sum for the file and store it. Use `shasum -c file.txt.sha256` to verify that the hash is correct, and that the original file hasn't been tampered with since the sum was made.

**Using tcpdump**

Use `ip link` to check the primary network interface name.

Use `sudo tcpdump -i eth0` to fill your terminal with a constant stream of text as new packets are read, use `Ctrl+C` to stop.

Use `sudo tcpdump -i eth0 -vn` to enable verbose and disable reverse DNS lookup.

Use `dig @8.8.8.8 A [example.com](http://example.com/)` to query a specific DNS server (in this case 8.8.8.8), asking it for the `A record` for the specified domain (in this case "example.com").
